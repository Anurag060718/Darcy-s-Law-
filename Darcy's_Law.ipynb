```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Darcy law well bore radii vs pressure drawdown.png

```python
def pressureprofile():
    re = float(input('Outer radius of Reservoir(ft): '))
    rwf = float(input('Wellbore Radius(ft): '))
    pwf = float(input('Bottomhole Pressure(PSI): '))
    h = float(input('Net Pay Thickness(ft): '))
    k = float(input('Average Reservoir Permeability(mD): '))
    q = float(input('Flowrate(STB/Day): '))
    mu = float(input('Oil Viscosity: '))
    B = 1
    r = np.linspace(rwf,re,500)
    pressure = []
    for i in range(len(r)):
      p = pwf + (141.2*q*mu*B*(np.log(r[i]/rwf))/(k*h))
      pressure.append(p)
    plt.style.use( 'classic')
    plt.figure(figsize=[8,6])
    plt.plot(r,pressure)
    plt.xlabel('radius (ft)',size = 20)
    plt.ylabel('Pressure (psi)',size = 20)
    plt.title(' Reservoir Pressure Profile',size = 25)
    plt.grid(True)
```

```python
pressureprofile()
```
